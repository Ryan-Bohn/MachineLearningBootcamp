```
ratingsWithoutColumns = ratingsDF.drop(columns = ["userId", "timestamp"]).groupby("movieId").mean()
ratingsWithoutColumns
```

```
moviesWithoutTitles = moviesDF.drop(columns = ["title"])
moviesWithoutTitles
```

```
cominedDF = pd.concat([moviesWithoutTitles, ratingsWithoutColumns], axis = 1, join = "inner")
cominedDF
```

```
joinedDF = pd.merge(moviesWithoutTitles, ratingsWithoutColumns, left_on = "movieId", right_on = "movieId")
#pd.merge(moviesWithoutTitles, ratingsWithoutColumns, left_on = "movieId", right_on = "movieId")
#moviesWithoutTitles.set_index("movieId").join(ratingsWithoutColumns.set_index("movieId"))
joinedDF
```

```
genres = joinedDF["genres"].str.get_dummies(sep = "|")
genres
```
