###Indexing

Python is a zero-indexed language!

```
a = "Hungry"
a[0]
```

```
a[1]
```

You can also count from the back!

```
a[-1]
```

###Slicing

Note the string only goes to the second index in slicing - 1.

```
a[1:3]
```

Skipping letters

1st digit represents starting index, 2nd digit represents ending index(not inclusive), 3rd digit represents skipped digits

```
a = "pneumonoultramicroscopicsilicovolcanoconiosis"
a[0:14:2]
```

Any of the digits can be ommitted

Default value for 1st digit - 0

Default value for 2nd digit - len(string)

Default value for 3rd digit - 1

```
a[::]
```

###Length of String

All characters, spaces and punctuation are counted.

```
a = "Let's go!"
len(a)
```

###Methods for Strings



Replace

```
b = "fellow"
print(b.replace("f","m"))
```

Variable does not change when method is applied

```
print(b)
```

```
b = b.replace("f","m")
print(b)
```

Split

```
closing = "Best Regards, Sincerely, From"
closing.split(",")
```

###Check String

```
a = "bye"
b = "bye felicia"
print (a in b)
```

Not In

```
a = "chair"
b = "car"
print (a not in b)
```
