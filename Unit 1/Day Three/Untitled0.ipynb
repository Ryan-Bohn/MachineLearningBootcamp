```
def manyArgs(*ar):
  for i in range(0, len(ar), 2):
    print(ar[i])

manyArgs("string", "tuple", "int", "float", "list")
```

```
def fibonacci(n) :
  if n <= 0 or type(n) != int:
    return "N/A";
  elif n == 1:
    return 0
  elif n == 2:
    return 1
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)
```
