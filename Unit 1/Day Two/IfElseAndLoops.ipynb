Lambda Math

```
#filter

numbers = [-1, 0, -1111, 3, 6]

positives = list(filter(lambda x: x>0, numbers))
positives
```

```
#map

numbers = [-1, 0, -1111, 3, 6]

mult3 = list(map(lambda x: x * 3, numbers))
mult3
```
