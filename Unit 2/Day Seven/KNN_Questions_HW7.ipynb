###1-D example for 1 nearest neighbor

The objective of this problem is to find the classification using 1-NN (1 nearest neighbor)

The training dataset consists of blue and red points. We want to find the classification of the black point. Is it a blue point or a red point?

Note: Blue represents +1 and red is -1.

```
plt.plot([1, -4, 5, 6], [0, 0, 0, 0], 'ro')
plt.plot([-2,2,3], [0, 0, 0],'bo')
plt.plot([0], [0], 'ko')
plt.show()
```

Graphically speaking, what is the classification of the black point?

Red point

```
training_data = [1, -2, -4, 2, 5, 3, 6]
training_labels = [-1, 1, -1, 1, -1, 1, -1]
testing_point = 0
```

Write a function that takes in the following parameters - training data, training labels, testing data and outputs the predicted label for the testing data.

Hint: Identify the closest point in training data list to testing data. The testing data will take the label of the nearest data point. Identify the closest point in training data list to testing data. The testing data will take the label of the nearest data point.

Hint: You can implement this anyway you want using concepts you have learned in class or additional methods found in documentation. Some things that could potentially help you are for loops, map function, etc.

```
def find_nn_1d(training_data, training_labels, testing_point):
  subtracted = list(map(lambda x: abs(x-testing_point), training_data))
  closest_point = min(subtracted)
  closest_indices = [index for index, value in enumerate(subtracted) if value == closest_point]
  #Same as:
  #for index, value in enumerate(subtracted):
  #  if value == closest_point:
  #    closest_indices.append(index)
  closest_labels = [training_labels[index] for index in closest_indices]
  labels = set(closest_labels)
  if len(labels) > 1:
    return "Inconclusive"
  return labels.pop()

find_nn_1d(training_data, training_labels, testing_point)
```

###2-D Example for 1 nearest neighbor

```
training_data = np.array([[1, -2, -4, 2, 5, 3, 6],[1, 1, 1, 1, 1, 1, 1]]).T
training_labels = np.array([[-1, 1, -1, 1, -1, 1, -1]]).T
testing_point = np.array([[0],[0]])

print(training_data)
print(training_labels)
print(testing_point)
```

Write a function that takes in the following parameters - training data, training labels, testing data and outputs the predicted label for the testing data.

Hint: Use numpy, avoid for loops as the plague

Hint: Distance should be euclidean

Hint: https://stackoverflow.com/questions/4370975/python-numpy-euclidean-distance-calculation-between-matrices-of-row-vectors may be helpful

```
from math import sqrt
def find_nn_2d(training_data, training_labels, testing_point):
  euclidean_data = (training_data - testing_point.T) ** 2
  euclidean_data = np.sum(euclidean_data, axis = 1)
  euclidean_data = np.sqrt(euclidean_data)
  indices = np.where(euclidean_data == min(euclidean_data))
  closest_labels = training_labels[indices]
  if np.all(closest_labels == closest_labels[0]):
    return closest_labels[0][0]
  return "Inconclusive"

find_nn_2d(training_data, training_labels, testing_point)
```
